# 1️⃣ strip()을 함께 써야 하는 이유 (sys.stdin.readline() 기준)

###⭐ 개념 요약
- sys.stdin.readline()으로 입력을 받으면 문자열 끝에 개행문자 \n이 자동으로 포함된다.
- 이 상태에서는 isdigit(), 문자열 비교 등이 의도와 다르게 동작할 수 있다.
- strip()은 문자열의 앞뒤 공백과 개행문자(\n)를 제거해준다.

###⭐ strip()을 안 써도 지금까지 에러가 안 났던 이유
- sys.stdin.readline()은 문자열 끝에 개행문자 \n을 포함하지만,
- int()나 split()을 함께 사용하면 이 개행이 자동으로 처리되기 때문에
- strip() 없이도 문제가 발생하지 않았다.

- int() → 앞뒤 공백 및 \n 자동 무시

- split() → 공백·개행을 기준으로 분리 (\n이 결과에 남지 않음)

###⭐ input()은 왜 문제가 없어 보이는가

- input()은 내부적으로 rstrip('\n')이 적용되어 있다.

- 그래서 개행이 없는 것처럼 보인다.

###⭐ 결론적으로 단일 문자열을 입력받을 때는
- s = sys.stdin.readline().strip()

###⭐ input() 과 sys.stdin.readline()의 차이

- input()은 입력을 요청하는 함수 → 입력창이 뜬다.

- sys.stdin.readline()은 표준입력을 읽기만 하는 함수 → 입력창이 안 뜬다.

# 2️⃣ get()

### ⭐ get() 함수는 딕셔너리에서 키(key)에 해당하는 값을 안전하게 가져오는 방법입니다.
```python
dict.get(key, default)
```
- key : 찾고 싶은 키
- default : 해당 키가 없을 때 반환할 값 (생략 시 None)


### ⭐ 일반 접근 방식의 문제점
```phthon
dic[i] = dic[i] + 1
```
- 이 코드는 i가 딕셔너리에 이미 존재할 때만 정상 동작합니다.
- 처음 등장하는 값이라면 KeyError가 발생합니다.

### ⭐ get()을 사용한 안전한 방식
```python
dic[i] = dic.get(i, 0) + 1
```

- 만약 i가 이미 딕셔너리에 있으면 → 기존 값 반환

- 만약 i가 없으면 → 0을 반환

- 그 값에 +1을 해서 다시 저장

# 3️⃣ str.isdigit()

### ⭐ 개념 요약

- 문자열이 모두 숫자로만 이루어져 있는지 판별

- 문자열 메서드이므로 int에는 직접 사용 불가

- 하나라도 숫자가 아니면 False
- 음수 표시인   -가 있어도 False

````python
print("123".isdigit())
````

````python
print("001".isdigit())
````

````python
print("12a".isdigit())
````

````python
print("-3".isdigit())
````

````python
print("3.14".isdigit())
````

````python
print("3 4".isdigit())
````

- 딕셔너리 만들기 연습문제 예제

````python
arr = [3, 1, 4, 0, 9, 2, 2, 4, 2, 0]

count = {}

for i in arr:
    count[i] = count.get(i, 0) + 1

print(count)
````
