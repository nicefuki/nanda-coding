# 파이썬 Set(집합) 명령어 정리
코딩테스트와 실무에서 가장 자주 사용하는 **핵심 명령어 중심**으로 정리

## 1. Set 기본 특성
- 중복 허용 X
- 순서 없음(인덱싱 불가)
- 빠른 탐색 성능(`in` 연산 O(1))
- set() / update() → ( ) 안에 iterable만 가능



````
# 빈 set 만들기

s = set()
print(s)
````

````
# 순서 상관없고, 중복은 안됨

s = {2, 3, 0, 1, 2, 5}
print(s)
````

````
# 리스트를 set으로

s = set([1, 2, 3, 1, 4, 0])
print(s)
````

````
# 튜플을 set으로

s = set((3, 2, 1, 1, 0))
print(s)
````

````
# 딕셔너리를 set으로

s = set({'a': 1, 'b': 2, 'c': 3})
print(s)
````

````
# 문자열을 set으로

s = set('hello')
print(s)
````

## 2. 원소 추가/삭제
```python
s.add(10)           # 하나 추가
s.update([4, 5, 6]) # 여러 개 추가

s.remove(3)         # 없으면 에러
s.discard(3)        # 없어도 에러 X

s.clear()           # 전체 삭제
```

````
s = {1, 2, 3}

s.add(4)
print(s)
````

````
s.update([5, 6, 7])
print(s)
````

````
s.update((8, 9))
print(s)
````

````
s.remove(9)
print(s)
````

````
s.remove(9)
print(s)
````

````
s.discard(9)
print(s)
````

````
s.discard(8)
print(s)
````

````
s.clear()
print(s)
````

## 3. 집합 연산
```python
a | b  # 합집합
a & b  # 교집합
a - b  # 차집합
a ^ b  # 대칭 차집합
```

````
a = {1, 2, 3, 4, 5}
b = {3, 4, 5, 6, 7, 8}

print(a | b)
````

````
print(a & b)
````

````
print(a - b)
````

````
print(a ^ b)
````

## 4. 집합 관계 확인 함수
```python
a.issubset(b)     # 부분집합
a.issuperset(b)   # 상위집합
a.isdisjoint(b)   # 공통 원소 없음
```

````
a = {1, 2, 3, 4}
b = {1, 2}
print(a.issubset(b))
print(b.issubset(a))
````

````
a = {1, 2, 3, 4}
b = {1, 2}
print(a.issuperset(b))
print(b.issuperset(a))
````

````
a = {1, 2, 3, 4}
b = {1, 2}
print(a.isdisjoint(b))
````

````
a = {4}
b = {1, 2}
print(a.isdisjoint(b))
````

## 5. Set 정렬하기 (중요)

set은 **순서가 없는(unordered)** 자료구조라서  
요소를 정렬하려면 반드시 `sorted()`를 사용해야 합니다.

```python
# 오름차순 정렬 (리스트 반환)
result_asc = sorted(s)
print(result_asc)

# 내림차순 정렬
result_desc = sorted(s, reverse=True)
print(result_desc)

````
s = {2, 3, 0, 1, 5, 4}
a = sorted(s)
print(a)
````

````
s = {2, 3, 0, 1, 5, 4}
a = sorted(s, reverse=True)
print(a)
````

## 6. 성능 핵심 요약
- `in` 탐색은 set이 가장 빠름 (O(1))
- 합집합/교집합/차집합 역시 빠름
- 코딩테스트: **자료 존재 여부 체크는 무조건 set을 우선 고려**

## 7. frozenset

- 딕셔너리의 Key는 일반 set은 안 되고, frozenset은 된다.
- 딕셔너리의 Value는 일반 set도 되고, frozenset도 된다.
